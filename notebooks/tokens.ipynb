```python
import tiktoken
import re

# Sample text
text = "Pernambuco meu país."

# Characters (including spaces and punctuation)
char_count = len(text)

silabas = ["Per", "na", "m", "bu", "co", "meu", "pa", "ís"]

# Words (splitting by whitespace)
word_count = len(text.split())

# Tokenization using tiktoken (using GPT-3.5 encoding)
encoding = tiktoken.get_encoding("cl100k_base")
tokens = encoding.encode(text)
token_count = len(tokens)

print(f"Texto: {text}")
print(f"caracteres: {[c for c in text]}")
print(f"nº caracteres: {char_count}")
print(f"silabas: {silabas}")
print(f"nº silabas: {len(silabas)}")
print(f"palavras: {text.split()}")
print(f"nº palavras: {word_count}")
print(f"Tokens: {token_count}")
print("\nTokens decoded:")
for token in tokens:
    print(f"{token}: {encoding.decode([token])}")
```
